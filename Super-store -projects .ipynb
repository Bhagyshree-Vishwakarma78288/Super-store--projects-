# Superstore project

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_excel(r"Superstore_USA orignal.xlsx")
df.head(2)  
```

```python
df.shape  
```

```python
# find out missing value
```

```python
df.isnull().sum() 
```

```python
df.fillna(df["Product Base Margin"].mean(),inplace=True)  
```

```python
df.isnull().sum()  
```

# order priority

```python
df["Order Priority"].value_counts() 
```

```python
df["Order Priority"].unique() 
```

```python
df["Order Priority"] = df["Order Priority"].replace('Critical ','Critical')
df["Order Priority"]
```

```python
ax = sns.countplot(x="Order Priority",data=df)
for bars in ax.containers:
    ax.bar_label(bars) 
plt.figure(figsize=(5,3)) 
plt.title("count of Order Priority")
plt.savefig("count of Order Priority.jpg")
plt.show() 
```

# Ship Mode

```python
df["Ship Mode"].value_counts() 
```

```python
ax = sns.countplot(x="Ship Mode",data=df)  # bar chart
for bars in ax.containers:
    ax.bar_label(bars) 
plt.title("count of Ship mode")
plt.figure(figsize=(5,3))
plt.show() 
```

```python
x = df["Ship Mode"].value_counts().index
y = df["Ship Mode"].value_counts().values
```

```python
plt.pie(y,labels=x,autopct="%0.2f%%") 
plt.figure(figsize=(5,4))
plt.legend() 
plt.show() 
```

# ship mode and product cartegory

```python
ax = sns.countplot(x="Ship Mode",data=df,hue="Product Category")
for bars in ax.containers:
    ax.bar_label(bars)
plt.title("ship mode category wise")
plt.figure(figsize=(5,3))
plt.show()
    
```
